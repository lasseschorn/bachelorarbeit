```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy.special import expit as sigmoid  # logistic function
import variablen as var
import warnings
import utils_plot
```

```python
X_complete = pd.read_csv(var.original_df)
```

```python
# entfernen der ID column
X_complete =  X_complete.drop(columns=['ID'])
# iterating the columns 

for column in X_complete.columns:
    s = X_complete[column].drop_duplicates()
    s.to_csv(var.prepair_df_dir + 'columns/' + column  ,index=False,sep=';', decimal=',',  float_format='%.2f')
    df = pd.read_csv(var.prepair_df_dir + 'columns/' + column  , delimiter=";", decimal=",")
    for i, v in df.iterrows():
        X_complete[column] = X_complete[column].replace(v[column], i  )
    X_complete[column] = X_complete[column].astype('float64')
    
```

```python
X_complete.to_csv(var.prepair_df_dir + var.csv_dataset_name ,index=False,sep=';', decimal=',', float_format='%.2f')
```
